```python
def str2function_bio(function):
    function = function.replace('Exp', 'math.exp')
    
    function = function.replace('[p]', 'densi')
    function = function.replace('[v]', 'v')
    function = function.replace('[d130]', 'd130')
    function = function.replace('[ht]', 'ht')
    function = function.replace('LOG', 'math.log')    
    
    varibales = ["densi","v","d130","ht"]
    filtered_list = [char for char in varibales if char in function]

    function = eval(f'lambda {', '.join(filtered_list)}: {function}')
    return function,filtered_list
```

```python
def str2function_bio(function):
    function = function.replace('Exp', 'math.exp')
    
    function = function.replace('[p]', 'densi')
    function = function.replace('[v]', 'v')
    function = function.replace('[d130]', 'd130')
    function = function.replace('[ht]', 'ht')
    function = function.replace('LOG', 'math.log')    
    
    varibales = ["densi","v","d130","ht"]
    filtered_list = [char for char in varibales if char in function]

    function_ = eval(f'lambda {', '.join(filtered_list)}: {function}')
    #    print(f'lambda {', '.join(filtered_list)}: {function}')
    return function_,filtered_list
```

```python
for i in element_counts:
    f,var = str2function_bio(i)
    print(var)
```

```python
for index, row in df.iterrows():

    try:
        f,var = str2function_bio(row.biomasa_eq)
        if var == ['densi', 'd130', 'ht']:
            biomasa[index] = f(float(row.densidad_eq),row.diametro,row.altura)
        elif var == ['d130', 'ht']:
            biomasa[index] = f(row.diametro,row.altura)
        elif var == ['densi', 'v']:
            biomasa[index] = f(float(row.densidad_eq),row.volumen)  
        else:
            biomasa[index] = f(row.diametro)
    except:
        f,var = str2function_bio(row.biomasa_eq)
        
        print(index,var)

    
```

```python
def str2function_car(function):
    
    function = function.replace('[b]', 'b')
    function = function.replace('[d130]', 'd130')
    function = function.replace('[ht]', 'ht') 
    
    varibales = ["b","d130","ht"]
    filtered_list = [char for char in varibales if char in function]

    function = eval(f'lambda {', '.join(filtered_list)}: {function}')
    return function,filtered_list
```

```python
for i in counter:
    f,var = str2function_car(i)
    print(var)
```

```python
for index, row in df.iterrows():

    try:
        f,var = str2function_car(row.carbon_eq)
        if var == ['b']:
            carbono[index] = f(row.biomasa)
        elif var == ['d130']:
            carbono[index] = f(row.diametro)
        else:
            carbono[index] = f(row.diametro,row.altura)
    except:
        print(index)
    
```
